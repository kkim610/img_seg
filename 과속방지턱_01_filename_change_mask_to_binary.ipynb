```python
import os
import shutil
import numpy as np
from PIL import Image

#L=['test_img','test_mask','train_img','train_mask','val_img','val_mask']
read_dir="speed_hump/input/"
obj_name='speed_hump'

print("read_dir",read_dir)
dir_L = os.listdir(read_dir)
print("dir_L",dir_L)

save_dir="dataset3/"
print("save_dir",save_dir)

# Retrieve a list of all files in the folder
for dd in dir_L:
    dr_name=read_dir+dd+"/"
    print("dr_name",dr_name) 
    if not os.path.exists(dr_name):
        print("읽을 디렉토리 없음",dr_name) 
        continue
    
    save_dir_name = save_dir+dr_name

    print("save_dir_name",save_dir_name)    
    LL = os.listdir(dr_name)

    print("LL",LL)
    read_dir_mask = dr_name+"mask"
    print("read_dir_mask",read_dir_mask)
    files = os.listdir(read_dir_mask)
    files.sort()

    save_dir_mask =save_dir_name+"mask/"+obj_name
    if not os.path.exists(save_dir_mask):
        os.makedirs(save_dir_mask)
    print("save_dir_mask",save_dir_mask)

    read_dir_img = dr_name+"images"
    print("read_dir_img",read_dir_img)
    save_dir_img =save_dir_name+"images"
    if not os.path.exists(save_dir_img):
        os.makedirs(save_dir_img)
    print("save_dir_img",save_dir_img)

    j = 0
    k = 0
    
    for filename in files:

        fname = read_dir_mask+"/"+filename
        image = Image.open(fname)
        
        # Convert the image to a NumPy array
        Z = np.array(image)
        ZZ=Z[:,:,0]==0 

        count=np.sum(ZZ)

        #print(j, count)       
        if count < 262*3+1 :
            k += 1
            continue

        j += 1
        Z[Z==0] = 255
        Z[Z==19] = 0
        Z=Z.astype('ubyte')
        if j > 999:
            new_fname = "image"+str(j)+".png"
        elif j > 99:
            new_fname = "image0"+str(j)+".png"
        elif j > 9:
            new_fname = "image00"+str(j)+".png" 
        else:
            new_fname = "image000"+str(j)+".png" 
                
        im = Image.fromarray(Z)  
        save_new_fname = save_dir_mask+"/"+new_fname       
        im.save(save_new_fname)  

        shutil.copy(read_dir_img+"/"+filename, save_dir_img)

        os.rename(save_dir_img+"/"+filename, save_dir_img+"/"+new_fname)    
       
    print(k, j, k*100/(j+k))

print(zzz)
```
